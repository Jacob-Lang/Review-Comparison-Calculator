```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
p = 0.5
n = 50
k_scan = range(0,n+1)
n_reps = 10000
bd = [stats.binom.pmf(k, n, p) for k in k_scan]

bd_hist = stats.binom.rvs(n, p, size=n_reps) 

plt.plot(k_scan, bd)
plt.hist(bd_hist, density=True, bins=range(0,n+2), align='left');
```

```python
n = 10
k = 10

p_scan = np.linspace(0,1,1000)

#plt.plot(p_scan, Pp)
plt.plot(p_scan, stats.beta.pdf(p_scan, k+1,n-k+1))
plt.vlines((k+1)/(n+2),0,25)

n = 100
k = 95

#plt.plot(p_scan, Pp)
plt.plot(p_scan, stats.beta.pdf(p_scan, k+1,n-k+1))
plt.vlines((k+1)/(n+2),0,25)

n = 1000
k = 930

#plt.plot(p_scan, Pp)
plt.plot(p_scan, stats.beta.pdf(p_scan, k+1,n-k+1))
plt.vlines((k+1)/(n+2),0,25)
```

## This works:

```python
n = 30
r = 1.0
k = round(r*n)

dp = 0.00001  # need fine enough grain. 
p_scan = np.arange(0,1+dp,dp)

pdf_X = stats.beta.pdf(p_scan, k+1,n-k+1)
cdf_X = stats.beta.cdf(p_scan, k+1,n-k+1)

n = 20
r = 1.0
k = round(r*n)

pdf_Y = stats.beta.pdf(p_scan, k+1,n-k+1)
cdf_Y = stats.beta.cdf(p_scan, k+1,n-k+1)

plt.plot(p_scan, pdf_X)
plt.plot(p_scan, pdf_Y)
plt.xlabel('Rating')
plt.title('Probability density of true rating given the reviews')
plt.yticks([]);

prob_XgtY = np.trapz(cdf_Y*pdf_X, dx=dp)
print("Probability that X is greater than Y:", round(prob_XgtY*100,2), "%")
```
